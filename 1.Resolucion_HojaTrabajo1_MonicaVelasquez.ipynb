### Resolución problema No. 4

```python
def plot_histograms(image_path):
    img = cv2.imread(image_path)
    
    if img is None:
        raise 
    
    b, g, r = cv2.split(img)
    
    # Escala de grises con promedio aritmético
    gray = np.mean(img, axis=2).astype(np.uint8)

    # Creación de histogramas
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    colors = ('blue', 'green', 'red')
    channels = (b, g, r)
    titles = ('Histograma Azul', 'Histograma Verde', 'Histograma Rojo', 'Histograma Escala de Grises')

    for ax, color, channel, title in zip(axes.flatten()[:3], colors, channels, titles[:3]):
        hist = cv2.calcHist([channel], [0], None, [256], [0, 256])
        ax.plot(hist, color=color)
        mean_val = np.mean(channel)
        ax.axvline(mean_val, color=color, linestyle='dashed', linewidth=1)
        ax.set_title(title)
        ax.set_xlim([0, 256])
        ax.legend([f'Mean: {mean_val:.2f}'])
    
    # Plot histogram for grayscale
    ax = axes.flatten()[3]
    hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
    ax.plot(hist, color='gray')
    mean_val = np.mean(gray)
    ax.axvline(mean_val, color='gray', linestyle='dashed', linewidth=1)
    ax.set_title(titles[3])
    ax.set_xlim([0, 256])
    ax.legend([f'Mean: {mean_val:.2f}'])
    
    plt.tight_layout()
    plt.show()
```

```python
image_path = 'img4.jpg'
plot_histograms(image_path)
```

### Enfoque de escala de grises ponderado

Es una técnica utilizada para convertir una imagen en color a una imagen en escala de grises. En lugar de simplemente promediar los valores de los canales de color (rojo, verde y azul), se asignan diferentes pesos a cada canal según su contribución a la percepción del brillo por el ojo humano. Esta técnica produce una imagen en escala de grises que se ve más natural y refleja mejor cómo percibimos los colores en términos de brillo.

Los pesos comúnmente utilizados en la conversión ponderada son:
- Rojo (R): 0.299
- Verde (G): 0.587
- Azul (B): 0.114

Estos pesos se basan en la sensibilidad del ojo humano a diferentes colores, siendo el verde el más sensible, seguido del rojo y finalmente del azul.

La fórmula para calcular el valor en escala de grises 𝑌
Y es:

𝑌 = 0.299 × 𝑅 + 0.587 × 𝐺 + 0.114 × 𝐵

```python
def weighted_grayscale(image_path):
    img = cv2.imread(image_path)
    
    if img is None:
        raise
    
    b, g, r = cv2.split(img)
    
    gray = 0.299 * r + 0.587 * g + 0.114 * b
    gray = gray.astype(np.uint8)
    
    return gray

image_path = 'img5.jpg'
gray_image = weighted_grayscale(image_path)

cv2.imwrite('img5_gray.jpg', gray_image)
```

### Resolución problema No. 6

 Es un modelo de color que representa los colores de una manera que es más intuitiva para los seres humanos que el modelo RGB. HSV significa Hue (matiz), Saturation (saturación) y Value (valor o brillo). Cada componente en el espacio de color HSV tiene un significado específico:

6.1 Hue (H): El matiz representa el tipo de color y se mide en grados, de 0° a 360°. Cada valor de matiz corresponde a un color en el círculo cromático:
- 0° es rojo
- 60° es amarillo
- 120° es verde
- 180° es cian
- 240° es azul
- 300° es magenta

6.2 Saturation (S): La saturación indica la pureza del color y se mide en un porcentaje de 0% a 100%. Un 0% de saturación significa un color completamente desaturado (gris), mientras que un 100% de saturación representa un color completamente puro (sin mezcla de gris).

6.3 Value (V): El valor también conocido como brillo, este se refiere a la luminosidad del color y se mide en un porcentaje de 0% a 100%. Un 0% de valor es completamente negro, mientras que un 100% de valor es el color más brillante posible.
